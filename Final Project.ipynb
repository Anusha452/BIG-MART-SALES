```python
df_n = df.select_dtypes(include= [np.number])
q1 = df_n.quantile(0.25)
q3 = df_n.quantile(0.75)
iqr = q3 - q1

lb = q1 - (1.5*iqr)
ub = q3 + (1.5*iqr)
outliers = df_n[(df_n < lb) | (df_n > ub)]
```

```python
outliers.count()
```

```python
for col in df_n.columns:
    col_median = df_n[col].median()
    df_n.loc[df_n[col] < lb[col], col] = col_median
    df_n.loc[df_n[col] > ub[col], col] = col_median
```

```python
outliers_ai = df_n[(df_n < lb) | (df_n > ub)]
```

```python
outliers_ai.count()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
cat_col = ['Item_Fat_Content','Outlet_Size','Outlet_Location_Type','Outlet_Type']
```

```python
for col in cat_col:
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
df
```

```python
df.drop(columns = ['Item_Identifier','Item_Type','Outlet_Identifier','Outlet_Establishment_Year'], inplace = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
x = df.drop(columns='Item_Outlet_Sales')
y = df['Item_Outlet_Sales']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state= 10)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape , y_test.shape
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(x.corr(),annot=True)
plt.show()
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
col=x.columns
col
```

```python
x_fit=s.fit_transform(x)
x_fit
```

```python
x=pd.DataFrame(x_fit,columns=col)
x
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
LR.fit(X_train, y_train)
```

```python
LR.intercept_
```

```python
LR.coef_
```

```python
LR.score(X_train,y_train)
```

```python
LR.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores=cross_val_score(LR,x,y,cv=5)
print(scores)
cv_score=scores.mean()
print(cv_score)
```

```python
p_values = LR.predict(X_test)
p = pd.DataFrame({'actual_value': y_test, 'predicted_value': p_values})
p
```

```python
from sklearn.metrics import r2_score
```

```python
r_value1= r2_score(y_test, p_values)
r_value1
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor()
RF.fit(X_train, y_train)
```

```python
scores=cross_val_score(RF,x,y,cv=5)
print(scores)
cv_score=scores.mean()
print(cv_score)
```

```python
p_values = RF.predict(X_test)
p = pd.DataFrame({'actual_value': y_test, 'predicted_value': p_values})
p
```

```python
r_value2= r2_score(y_test, p_values)
r_value2
```
